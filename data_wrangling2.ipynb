```python
import pandas as pd
import numpy as np
import json
```

```python
#load the JSON dataset into a pandas dataframe
businessJSON_Path='yelp_dataset/business.json'
businessdf = pd.read_json(businessJSON_Path,encoding='utf-8-sig', lines=True)
```

```python
list(businessdf.city)
```

```python
businessdf.categories.unique()
#I need to unpack the categories since they are comma separated and each string is acting as a separate value. They need to be arrays so I can unpack them.
```

```python
businessdf=businessdf.filter(items=['business_id', 'name','city', 'state', 'categories'])
#first keep only the relevant columns
```

```python
#dfCategories.categories.str.get_dummies(sep=', ')
#Chunk dataframe into one that's smaller
n = 10000  #chunk row size
list_df = [businessdf[i:i+n] for i in range(0,businessdf.shape[0],n)] #found function online to chunk up a df
```

```python
#break up the categories column
for chunk_index,chunk in enumerate(list_df):
    list_df[chunk_index]=pd.concat([list_df[chunk_index],list_df[chunk_index].categories.str.get_dummies(sep=', ')],1)
```

```python
#get only the restaurants
for chunk_index in range(len(list_df)):
    list_df[chunk_index]=list_df[chunk_index][list_df[chunk_index]['Restaurants']==1]
```

```python
#concat them all into one dataset
restaurant_df=pd.concat(list_df, sort=False)
```

```python
list(restaurant_df.columns)
```

```python
#Export to CSV so I don't have to always keep manipulating the JSON to get the dataframe
restaurant_df.to_csv('yelp_dataset/restaurant.csv',index=None,header=True, encoding="utf-8-sig") 
```
